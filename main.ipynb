```python
import pandas as pd
import glob
from IPython.display import display_html
from IPython.display import Markdown as md

file_list = [f for f in glob.glob("*.csv")]
file_list.sort(reverse=True)
latest_file = file_list[:3]
trust_csv, foreign_csv, dealer_csv = latest_file
df_foreign = pd.read_csv(foreign_csv)
df_trust = pd.read_csv(trust_csv)
df_dealer = pd.read_csv(dealer_csv)
date_string = latest_file[0].split('.')[0].split('_')[0]
date_string_f = f'{date_string[:4]}-{date_string[4:6]}-{date_string[6:]}'
md("# {} 三大法人買賣超排行".format(date_string_f))
```

### 外資

```python
df_top_styler = df_foreign.head(20).style.hide(
    axis='index').set_table_attributes("style='display:inline'").set_caption('買超排行')
df_bottom_styler = df_foreign.tail(20).sort_values(by='外陸資買賣超股數(不含外資自營商)').style.hide(
    axis='index').set_table_attributes("style='display:inline'").set_caption('賣超排行')

display_html(df_top_styler._repr_html_() +
             df_bottom_styler._repr_html_(), raw=True)
```

### 投信

```python
df_top_styler = df_trust.head(20).style.hide(
    axis='index').set_table_attributes("style='display:inline'").set_caption('買超排行')
df_bottom_styler = df_trust.tail(20).sort_values(by='投信買賣超股數').style.hide(
    axis='index').set_table_attributes("style='display:inline'").set_caption('賣超排行')

display_html(df_top_styler._repr_html_() +
             df_bottom_styler._repr_html_(), raw=True)
```

### 自營商

```python
df_top_styler = df_dealer.head(20).style.hide(
    axis='index').set_table_attributes("style='display:inline'").set_caption('買超排行')
df_bottom_styler = df_dealer.tail(20).sort_values(by='自營商買賣超股數').style.hide(
    axis='index').set_table_attributes("style='display:inline'").set_caption('賣超排行')

display_html(df_top_styler._repr_html_() +
             df_bottom_styler._repr_html_(), raw=True)
```
